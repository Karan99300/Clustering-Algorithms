```python
import numpy as np
X=df.values
num_clusters = 3
num_crows = 50
num_dimensions = 13
max_ranges= df.max().values
max_iterations = 100
awareness_probability = np.random.rand()
best_crow=optimize_crows(X, max_iterations,awareness_probability,max_ranges,num_crows,num_clusters,num_dimensions)
print(best_crow)
```

```python
def initialize_crows_positions(num_crows, num_clusters, num_dimensions, max_ranges):
    positions = np.random.uniform(low=0, high=1, size=(num_crows, num_clusters, num_dimensions))
    print(positions.shape)

    for dim in range(num_dimensions):
        positions[:, :, dim] *= max_ranges[dim]

    return positions

initialize_crows_positions(num_crows, num_clusters, num_dimensions, max_ranges)
```
